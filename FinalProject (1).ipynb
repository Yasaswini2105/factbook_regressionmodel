```
import pandas as pd
data = pd.read_csv("factbook.csv", sep = ';')
data = data[1:]
performance_Based_on_GDP = data.iloc[:,9:10].values
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = "NaN",strategy = "mean",axis = 0)
imputer = imputer.fit(performance_Based_on_GDP)
performance_Based_on_GDP = imputer.transform(performance_Based_on_GDP)
data['GDP'] = performance_Based_on_GDP

import matplotlib.pyplot as plt
import pandas as pd

#data.nlargest(10,'GDP').plot(kind = 'line',x = 'Country',y='Area(sq km)',ax = ax)
data.nlargest(10,'GDP').plot(kind = 'bar',x = 'Country',y='GDP',figsize = (20,10),title = "PERFORMANCE OF COUNTRIES BASED OF GDP")
del data['HIV/AIDS - adult prevalence rate(%)']
del data['Internet hosts']
del data['Internet users']
del data['HIV/AIDS - deaths']
del data['HIV/AIDS - people living with HIV/AIDS']
del data['Natural gas - consumption(cu m)']
del data['Natural gas - exports(cu m)']
del data['Natural gas - imports(cu m)']
del data['Natural gas - proved reserves(cu m)']
del data['Natural gas - production(cu m)']
del data['Oil - consumption(bbl/day)']
del data['Oil - exports(bbl/day)']
del data['Oil - imports(bbl/day)']
del data['Oil - production(bbl/day)']
del data['Oil - proved reserves(bbl)']
del data['Telephones - main lines in use']
del data['Telephones - mobile cellular']
del data['Total fertility rate(children born/woman)']
del data['Railways(km)']
del data['Exports']
del data['Imports']
del data['Highways(km)']
#performance_Based_on_GDP = data.iloc[:,9:10].values
currentAccountBalance = data.iloc[:,3:4].values
imputer = imputer.fit(currentAccountBalance)
currentAccountBalance = imputer.transform(currentAccountBalance)
data['Current account balance'] = currentAccountBalance
AmmountOfDebtOwed = data.iloc[:,5:6].values
imputer = imputer.fit(AmmountOfDebtOwed)
AmmountOfDebtOwed = imputer.transform(AmmountOfDebtOwed)
data['Debt - external'] = AmmountOfDebtOwed
data.nlargest(10,'Current account balance').plot(kind = 'line',x = 'Country',y = 'Current account balance',figsize = (15,10), title = "10 countries with high current account balance")
birthRate = data.iloc[:,2:3].values
imputer = imputer.fit(birthRate)
birthRate = imputer.transform(birthRate)
data['Birth rate(births/1000 population)'] = birthRate
deathRate = data.iloc[:,4:5].values
imputer = imputer.fit(deathRate)
deathRate = imputer.transform(deathRate)
data['Death rate(deaths/1000 population)'] = deathRate
ElectricityConsumption = data.iloc[:,6:7].values
imputer = imputer.fit(ElectricityConsumption)
ElectricityConsumption = imputer.transform(ElectricityConsumption)
data['Electricity - consumption(kWh)'] = ElectricityConsumption
ElectricityProduction = data.iloc[:,7:8].values
imputer = imputer.fit(ElectricityProduction)
ElectricityProduction = imputer.transform(ElectricityProduction)
data['Electricity - production(kWh)'] = ElectricityProduction

GDP_perCapita = data.iloc[:,9:10].values
imputer = imputer.fit(GDP_perCapita)
GDP_perCapita = imputer.transform(GDP_perCapita)
data['GDP - per capita'] = GDP_perCapita
GDP_realGrowthRate = data.iloc[:,10:11].values
imputer = imputer.fit(GDP_realGrowthRate)
GDP_realGrowthRate = imputer.transform(GDP_realGrowthRate)
data['GDP - real growth rate(%)'] = GDP_realGrowthRate
IndustrialProductionRate = data.iloc[:,11:12].values
imputer = imputer.fit(IndustrialProductionRate)
IndustrialProductionRate = imputer.transform(IndustrialProductionRate)
data['Industrial production growth rate(%)'] = IndustrialProductionRate
InfantMortalityRate = data.iloc[:,12:13].values
imputer = imputer.fit(InfantMortalityRate)
InfantMortalityRate = imputer.transform(InfantMortalityRate)
data['Infant mortality rate(deaths/1000 live births)'] = InfantMortalityRate
InflationRate = data.iloc[:,13:14].values
imputer = imputer.fit(InflationRate)
InflationRate = imputer.transform(InflationRate)
data['Inflation rate (consumer prices)(%)'] = InflationRate
InvestmentGrossFixed = data.iloc[:,14:15].values
imputer = imputer.fit(InvestmentGrossFixed)
InvestmentGrossFixed = imputer.transform(InvestmentGrossFixed)
data['Investment (gross fixed)(% of GDP)'] = InvestmentGrossFixed
LabourForce = data.iloc[:,15:16].values
imputer = imputer.fit(LabourForce)
LabourForce = imputer.transform(LabourForce)
data['Labor force'] = LabourForce
LifeExpectancy = data.iloc[:,16:17].values
imputer = imputer.fit(LifeExpectancy)
LifeExpectancy = imputer.transform(LifeExpectancy)
data['Life expectancy at birth(years)'] = LifeExpectancy
MilitaryExpenditure = data.iloc[:,17:18].values
imputer = imputer.fit(MilitaryExpenditure)
MilitaryExpenditure = imputer.transform(MilitaryExpenditure)
data['Military expenditures - dollar figure'] = MilitaryExpenditure
MilitaryExpenditureGDP = data.iloc[:,18:19].values
imputer = imputer.fit(MilitaryExpenditureGDP)
MilitaryExpenditureGDP = imputer.transform(MilitaryExpenditureGDP)
data['Military expenditures - percent of GDP(%)'] = MilitaryExpenditureGDP
population = data.iloc[:,19:20].values
imputer = imputer.fit(population)
population = imputer.transform(population)
data['Population'] = population
publicDebt = data.iloc[:,20:21].values
imputer = imputer.fit(publicDebt)
publicDebt = imputer.transform(publicDebt)
data['Public debt(% of GDP)'] = publicDebt
ReservesOfForeignExchange = data.iloc[:,21:22].values
imputer = imputer.fit(ReservesOfForeignExchange)
ReservesOfForeignExchange = imputer.transform(ReservesOfForeignExchange)
data['Reserves of foreign exchange & gold'] = ReservesOfForeignExchange
unemploymentRate = data.iloc[:,22:23].values
imputer = imputer.fit(unemploymentRate)
unemploymentRate = imputer.transform(unemploymentRate)
data['Unemployment rate(%)'] = unemploymentRate
data
```
